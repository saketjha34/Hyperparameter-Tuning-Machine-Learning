```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv\\heart.csv')
df
```

```python
input_cols = ['age', 'sex', 'chest pain type', 'resting bp s', 'cholesterol',
       'fasting blood sugar', 'resting ecg', 'max heart rate',
       'exercise angina', 'oldpeak', 'ST slope',]
target_col = 'target'
numeric_cols = ['age', 'sex', 'chest pain type', 'resting bp s', 'cholesterol',
       'fasting blood sugar', 'resting ecg', 'max heart rate',
       'exercise angina', 'oldpeak', 'ST slope',]
```

```python
train_inputs = df[input_cols].copy()
train_targets = df[target_col].copy()
```

```python
train_inputs
```

```python
train_targets
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train_inputs[numeric_cols])

train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
train_inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, train_targets, val_targets = train_test_split(train_inputs, train_targets, test_size=0.35  ,random_state=42)
```

## Logistic regression

```python
lr = LogisticRegression()
lr.fit(X_train,train_targets)
lr.score(X_val,val_targets)
```

## DECISION TREE

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,train_targets)
dt.score(X_val,val_targets)
```

## RANDOM FOREST

```python
rf = RandomForestClassifier()
rf.fit(X_train,train_targets)
rf.score(X_val,val_targets)
```

```python
def get_score(model, X_train, X_val, train_targets, val_targets):
    model.fit(X_train, train_targets)
    return model.score(X_val, val_targets)
```

## Parameter tunning using k fold cross validation

```python
scores1 = cross_val_score(RandomForestClassifier(n_estimators=5),X_train,train_targets,cv=10)
np.average(scores1)
```

```python
scores2 = cross_val_score(RandomForestClassifier(n_estimators=50),X_train,train_targets,cv=10)
np.average(scores2)
```

```python
scores2 = cross_val_score(RandomForestClassifier(n_estimators=100,max_depth=7),X_train,train_targets,cv=10)
np.average(scores2)
```

```python
for i in range(1,11):
    scores = cross_val_score(RandomForestClassifier(n_estimators=100,max_depth=i),X_train,train_targets,cv=10)
    print(f'max depth - {i}',np.average(scores))
```

```python
for i in range(1,201,20):
    scores = cross_val_score(RandomForestClassifier(n_estimators=i,max_depth=10),X_train,train_targets,cv=10)
    print(f'n_estimators:{i} average_scores : {np.average(scores)}')
```

```python
scores = cross_val_score(RandomForestClassifier(n_estimators=141,max_depth=10,max_leaf_nodes=127 , n_jobs=-1, random_state=42, max_samples=0.97, criterion='gini'),X_train,train_targets,cv=10)
np.average(scores)
```

```python
best_params_rf = {
    'random_state':42,
    'n_estimators':141,
    'max_depth':10,
    'max_leaf_nodes':127 ,
    'n_jobs':-1, 
    'random_state':42,
    'max_samples':0.97, 
    'criterion':'gini',
}
```

```python
best_model = RandomForestClassifier(**best_params_rf).fit(X_train,train_targets)
print(f'train_score - {round(best_model.score(X_train,train_targets)*100,2)}%')
print(f'val_score - {round(best_model.score(X_val,val_targets)*100,2)}%')
```
